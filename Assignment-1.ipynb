## Q3. How does Python support multiple programming paradigms, and what are they?

Python's ability to support multiple programming paradigms—procedural, object-oriented, functional, and declarative—offers developers a versatile toolkit for solving diverse problems. This flexibility allows developers to choose the most suitable paradigm for a given task, enhancing productivity and code quality.

- Procedural Programming in Python

Procedural programming organizes code into functions and executes it sequentially. Python's straightforward syntax makes it ideal for procedural tasks, such as scripting or automating repetitive processes. For example, a simple function to sum a list demonstrates Python's procedural capabilities:

def sum_list(numbers):
total = 0
for num in numbers:
total += num
return total

print(sum_list([1, 2, 3, 4])) # Output: 10
Copy
This paradigm is particularly useful for tasks requiring step-by-step instructions, such as data processing pipelines.

- Object-Oriented Programming (OOP) in Python

Python's OOP features, such as encapsulation, inheritance, and polymorphism, enable developers to model real-world entities effectively. OOP is ideal for building modular, reusable, and maintainable code. For instance:

class Employee:
def __init__(self, name, age):
self.name = name
self.age = age

def display_info(self):
print(f"{self.name} is {self.age} years old.")

emp = Employee("Alice", 30)
emp.display_info() # Output: Alice is 30 years old
Copy
This paradigm is well-suited for large-scale applications like web frameworks or game development.

- Functional Programming in Python

Python supports functional programming through lambda functions, higher-order functions, and tools like map, filter, and reduce. Functional programming emphasizes immutability and pure functions, making it ideal for tasks like data transformations:

from functools import reduce

numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result) # Output: 10
Copy
This paradigm is particularly effective for parallel processing and mathematical computations.

- Declarative Programming in Python

Declarative programming focuses on what to do rather than how to do it. Python's libraries, such as SQLAlchemy for database queries or TensorFlow for machine learning, allow developers to write declarative code. For example:

from sqlalchemy import Table, Column, Integer, String, MetaData

metadata = MetaData()
users = Table('users', metadata,
Column('id', Integer, primary_key=True),
Column('name', String),
Column('age', Integer))

## Q13. Explain the concept of Python bytecode and its role in Python's interpretive nature.

Python bytecode is an intermediate representation of your Python code that the interpreter generates before execution. It plays a central role in Python’s interpretive nature, acting as the bridge between human-readable source code and machine-level execution.
When you run a Python program, the CPython interpreter first compiles the source code (.py files) into bytecode
Bytecode is a low-level, platform-independent set of instructions stored in .pyc files (inside the __pycache__ directory)
This bytecode is then executed by the Python Virtual Machine (PVM).

### Role in Python's Interpretive Nature
- Interpretation process: Python doesn’t directly translate source code into machine code. Instead, it uses bytecode as an intermediate step, which the PVM interprets line by line.
- Portability: Since bytecode is platform-independent, the same Python code can run on Windows, macOS, or Linux without modification.
- Flexibility: This interpretive model allows Python to remain highly dynamic, supporting features like introspection, dynamic typing, and runtime modifications.
- Ease of debugging: Bytecode makes Python easier to analyze and debug compared to compiled languages.


## Q14. What is the Python Virtual Machine (PVM), and how does it execute Python programs?

### Python Virtual Machine/Python Interpreter
An interpreter is a kind of program that executes other programs. When you write a python program.
The python interpreter reads your program and carries out the instructions it contains.
In effect, the interpreter is a layer of software logic between your code and the computer hardware on your machine.
The python code you write must always be run by this interpreter. And to enable that, you must install a python interpreter on your computer.

### Program Execution
PVM is the runtime engine that executes Python bytecode. When you run a Python script, the source code is first compiled into bytecode, which is then interpreted by the PVM.
The PVM abstracts the underlying hardware and provides a consistent runtime environment for Python programs, allowing them to run on different platforms without modification.
PVM manages memory, handles garbage collection, and provides the necessary infrastructure for Python programs to execute.
Python code is compiled into bytecode (.pyc files) that is then executed by the PVM.
Compilation is simply a translation step, and byte code is a lower-level, platform-independent representation of your code.
Python translates each of your source statements into a group of byte code instructions by decomposing them into individual steps.
This byte code translation is performed to speed execution – byte code can be run much more quickly than the original source code statements in your text file.
Python byte code is not binary machine code (e.g., instructions for an intel or ARAM chip). Byte code is a python-specific representation.
Python will store the byte code of your programs in files that end with a .pyc extention (“.pyc” means compiled “.py” source.
In 3.2 and later, python instead saves its .pyc byte code files in a subdirectory named __pycache__ located in the directory where your source file reside, and in files whose names identify the python version that created them (e.g., script.cpython—33.pyc).
Python simply creates and uses the byte code in memory and discards it on exit.
To speed startups, through, it will try to save byte code in a file in order to skip the compile step next time around.
The next time you run your program, python will load the .pyc files and skip the compilation step, as long as haven’t changed your source code since the byte code was last saved, and aren’t running with a different python than the one that created the byte code.
Byte code compilation is automatic, and the PVM is just part of the python system that you have installed on your machine. Again, programmers simply code and run files of statements, and python handles the logistic of running them.
In python is runtime-there is no initial compile-time phase at all, and everything happens as the program is running.

## Q15. How does the PVM manage memory during the execution of a Python program?

- Memory Allocation Layers
    * OS Virtual Memory Manager (VMM): Provides raw memory pages to Python.
    * Python’s Object Allocator: Built on top of the OS allocator, optimized for Python’s object model.
    * Specialized Pools: Python maintains memory pools for small objects to reduce fragmentation and speed up allocation. 
- Reference Counting
    * Every Python object has a reference count.
    * When a variable points to an object, the count increases; when references are deleted, the count decreases.
    * If the count reaches zero, the object’s memory is immediately reclaimed.
- Garbage Collection (GC)
    * Handles cyclic references (objects referencing each other, preventing reference counts from dropping to zero).
    * Python’s GC periodically scans for unreachable cycles and frees them.
    * This hybrid model (reference counting + GC) balances responsiveness and memory safety.
- Memory Efficiency
    * Small Object Allocator (PyMalloc): Optimized for frequently used small objects (like integers, strings).
    * Interning: Common immutable objects (like short strings or integers) are reused to save memory.
    * Tiered Memory Management: Research shows Python VM can adapt memory tiers for large-scale applications, especially in data centers.

## Q30. Define literals in Python and provide examples of different types of literals.

In Python, literals are fixed values that are directly used in the code to represent data. They are not variables or expressions but rather the actual data itself. Literals are used to assign values to variables or as constants in the code.
Python supports various types of literals, including:
- Numeric literals: These represent numeric values and can be integers, floating-point numbers, or complex numbers. For example: 42, 3.14, 2 + 3j.
- String literals: These represent sequences of characters and can be enclosed in single quotes (') or double quotes ("). For example: 'Hello', "Python", "42".
- Boolean literals: These represent the truth values True and False.
- None literal: The None literal represents the absence of a value or a null value.
- Tuple literals: These represent immutable ordered collections of items enclosed in parentheses (()). For example: (1, 2, 3), ('a', 'b', 'c').
- Dictionary literals: These represent key-value pairs enclosed in curly braces ({}). For example: {'name': 'John', 'age': 30}.
- Set literals: These represent unordered collections of unique items enclosed in curly braces ({}). For example: {1, 2, 3}, {'apple', 'banana', 'orange'}

```python
name = 'Harsh'
age = 21
is_student = True
data = {'name': name, 'age': age, 'is_student': is_student}
print(data)
```
